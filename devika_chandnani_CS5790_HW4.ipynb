```python
import pandas as pd
import numpy as np
import sklearn
import skimage
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
img = skimage.io.imread('image.png') 
img = img.astype('uint8')/255 #Normalize the data
skimage.io.imshow(img)
plt.show()
```

```python
print(img.shape)
print(type(img))
#This part of the code convers the image into 2D and a float
pixel_vals = img.reshape((-1,3))
pixel_vals = np.float32(pixel_vals)
#shows us that img reassigned to pixel_vals is now a 2D array
print(pixel_vals.shape)
print(type(pixel_vals))
```

```python
def change_color(value):
    if value == 0:
        return (60, 179, 113) #SpringGreen
    elif value == 1:
        return (0, 191, 255) #DeepSkyBlue
    elif value == 2:
        return (255, 255, 0) #Yellow
    elif value == 3:
        return (255, 0, 0) #Red
    elif value == 4:
        return (0, 0, 0) #Black
    elif value == 5:
        return (169, 169, 169) #DarkGray
    elif value == 6:
        return (255, 140, 0) #Purple
    elif value == 7:
        return (128, 0, 128) #Pink
    elif value == 8:
        return (255, 192, 203) #White
    else:
        return (255, 255, 255)
```

```python
#When N = 2
n = 2
centroids2 = np.array([[0,0,0],[0.2,0.2,0.2]]) #Initialize specific centroid for N = 2
kmeans = KMeans(n_clusters=n, random_state=0,init=centroids2, max_iter=50, n_init=1) #max_iter parameter stops early if convergence is reached, or runs until 50
labels2 = kmeans.fit_predict(pixel_vals)
labels2 = pd.Series(labels2).map(lambda x: change_color(x))
labels2 = pd.DataFrame(labels2.to_list(), index = labels2.index)
k2 = np.array(labels2).reshape((244, 198,3))
SSE2 = round(kmeans.inertia_, 2) #picks final SSE
print("The final SSE for Kmeans when K=2 is", SSE2) #prints SSE
f, axarr = plt.subplots(1,2,figsize=(10,10))
axarr[0].set_xlabel('Original Image', fontsize=12)
axarr[1].set_xlabel('Segmented Image when K = 2', fontsize=12)  
axarr[0].imshow(img) #original image
axarr[1].imshow(k2) #use the 2D array to plot new image
plt.savefig("k4.png", dpi=125)
```
